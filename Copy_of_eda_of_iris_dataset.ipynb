```python
#EDA Iris dataset

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hotel_df = pd.read_excel("hotel_management_dataset.xlsx")
```

```python
hotel_df.head()
```

```python
#how many datapoints for each classes are present

hotel_df["Booking Source"].value_counts()

#this is a balanced dataset as each class has equal no of values or almost similar values is also balanced dataset(not imbalanced dataset)
```

```python
#2-d scatter plot

hotel_df.plot(kind='scatter',x='Payment Method',y='Booking Source')
plt.show()
```

```python
import seaborn as sns

sns.set_style("whitegrid")
sns.FacetGrid(hotel_df,hue="Room Type",height=4)\
    .map(plt.scatter,"Room Type","Check-in Date")\
    .add_legend()
plt.show()
```

```python
#plot on bases of petal


sns.set_style("whitegrid")
sns.FacetGrid(hotel_df,hue="Room Type",height=4)\
    .map(plt.scatter,"Room Number","Payment Method")\
    .add_legend()
plt.show()
```

```python
#Pair plot

plt.close();
sns.set_style("whitegrid")
sns.pairplot(hotel_df
,hue='Room Type', height=3);
plt.show()
```

```python
#Histogram(1-D sctter plot kind of)

sns.FacetGrid(hotel_df,hue="Room Type", height=5)\
    .map(sns.distplot, "Room Number")\
    .add_legend();
plt.show();
```

```python
#pdf and cdf

hotel=hotel_df[hotel_df['Guest Name']=='hotel']
print(hotel.head())
counts, bin_edges= np.histogram(hotel['Room Number'],bins=10, density= True)
pdf=counts/(sum(counts))
print(pdf)

print(bin_edges)

cdf=np.cumsum(pdf)
print(cdf)

plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show();
```

```python
#pdf and cdf

hotel=hotel_df[hotel_df['Guest Name']=='hotel']
print(hotel.head())
counts, bin_edges= np.histogram(hotel['Room Number'],bins=10, density= True)
pdf=counts/(sum(counts))
print(pdf)

print(bin_edges)

cdf=np.cumsum(pdf)
print(cdf)

plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)



hotel_ds=hotel_df[hotel_df['Guest Name']=='hotel_ds']

counts_ve, bin_edges_ve= np.histogram(hotel_ds['Room Number'],bins=10, density= True)
pdf_ve=counts_ve/(sum(counts_ve))
print(pdf_ve)

print(bin_edges_ve)

cdf_ve=np.cumsum(pdf_ve)
print(cdf_ve)

plt.plot(bin_edges_ve[1:],pdf_ve)
plt.plot(bin_edges_ve[1:],cdf_ve)


hotel_da=hotel_df[hotel_df['Check-in Date']=='hotel_da']

counts_vi, bin_edges_vi= np.histogram(hotel_da['Room Number'],bins=10, density= True)
pdf_vi=counts_vi/(sum(counts_vi))
print(pdf_vi)

print(bin_edges_vi)

cdf_vi=np.cumsum(pdf_vi)
print(cdf_vi)

plt.plot(bin_edges_vi[1:],pdf_vi)
plt.plot(bin_edges_vi[1:],cdf_vi)


plt.show();
```
